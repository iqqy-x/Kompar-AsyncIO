```python
import pandas as pd

df = pd.read_csv("data/raw/diskominfo-od_16357_kode_wilayah_dan_nama_wilayah_kecamatan_data.csv")
```

```python
# Hapus kecamatan tidak valid
df = df[df["kemendagri_kecamatan_nama"] != "KECAMATAN BELUM TERIDENTIFIKASI"]

# Hapus missing lat-long
df = df.dropna(subset=["latitude", "longitude"])
```

```python
# Pilih provinsi tertentu
df = df[df["kemendagri_provinsi_nama"] == "JAWA BARAT"]

# Normalisasi nama kolom
df = df.rename(columns={
    "kemendagri_kota_nama": "nama_kota",
    "kemendagri_kecamatan_nama": "nama_kecamatan",
})

# Kolom final yang dibutuhkan main.py
df_final = df[["nama_kota", "nama_kecamatan", "latitude", "longitude"]]
```

```python
# Simpan ke cleaned folder
df_final.to_excel("data/cleaned/kecamatan_cleaned.xlsx", index=False)

df_final.head()
```
